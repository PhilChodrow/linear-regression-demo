```python
from lab1.LinearRegression import LinearRegression1D

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split

np.random.seed(1234)
```

```python
url = "https://philchodrow.github.io/PIC16A/datasets/palmer_penguins.csv"
penguins = pd.read_csv(url)
gentoos = penguins[penguins["Species"].str.contains("Gentoo")]
```

```python
x = gentoos["Culmen Length (mm)"]
y = gentoos["Culmen Depth (mm)"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
plt.scatter(x_train, y_train)
```

```python
LR = LinearRegression1D(learning_rate = 0.00002, max_iter = 100, tol = 0.0001)
```

```python
LR.score(x, y)
```

```python
y_random_pred = LR.predict(x_train)
plt.scatter(x_train, y_train)

plt.plot(x_train, y_random_pred, color = "black")
```

```python
scores = LR.gradient_descent(x_train, y_train)
```

```python
y_pred = LR.predict(x_train)
plt.scatter(x_train, y_train)
plt.plot(x_train, y_pred, color = "black")
```

```python
LR.score(x_train, y_train)
```

```python
y_pred = LR.predict(x_test)
plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred, color = "black")
```

```python
LR.score(x_test, y_test)
```
